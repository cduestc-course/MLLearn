### 0-1标准化

```python
np.random.seed(24)
a = np.arange(24)
np.random.shuffle(a)
a = a.reshape(6, 4)
a
```

```python
# 计算每列最大值
a.max(axis = 0)
# np.max(a,0)
```

```python
# 计算每列最小值
a.min(axis = 0)
# np.min(a,0)
```

```python
# 计算每列极差
a.max(axis = 0) - a.min(axis = 0)
```

```python
# 对数据集a进行归一化计算
(a - a.min(axis = 0)) / (a.max(axis = 0) - a.min(axis = 0))
```

```python
def maxmin_norm(X):
    """
    max—min normalization标准化函数
    """
    maxmin_range = X.max(axis=0) - X.min(axis=0)
    return (X - X.min(axis=0)) / maxmin_range
```

```python
maxmin_norm(a)
```

### Z-score标准化

```python
np.random.seed(24)
a = np.arange(24)
np.random.shuffle(a)
a = a.reshape(6, 4)
a
```

```python
# 每列均值
a.mean(0)
```

```python
# 每列标准差
a.std(0)
```

```python
# Z-Score标准化
(a - a.mean(0)) / a.std(0)
```

```python
def z_score(X):
    """
    Z-Score标准化函数
    """
    return (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
z_score(a)
```
