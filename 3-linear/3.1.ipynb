```python
# 科学计算模块
import numpy as np
import pandas as pd

# 绘图模块
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# corrcoef
A = np.array([[1, 2, 3], [4, 5, 10]]).T
print(A)
np.corrcoef(A[:, 0], A[:, 1])
```

```python
# 取值为负
A = np.array([[1, 2, 3], [-1, -1.5, -5]]).T
plt.plot(A[:, 0], A[:, 1])
np.corrcoef(A[:, 0], A[:, 1])
```

```python
X = np.random.randn(20)
print(X)
y = X + 1

# 对应点图
plt.plot(X, y, 'o')
np.corrcoef(X, y)
```

```python
a = X.shape
print(a)
#np.random.normal?

# 创建一个和X一样的服从标准正态分布的随机数组
ran = np.random.normal(size = X.shape)
print(ran)
# delta控制扰动大小
delta = 0.5
# 计算扰动值
r = ran * delta
# 新的y值
y1 = y + r
```

```python
plt.subplot(121)
plt.plot(X, y, 'o')
plt.title('y=x+1')
plt.plot(X,y,'r-')
plt.subplot(122)
plt.plot(X, y1, 'o')
plt.title('y=x+1+r')
```

```python
np.corrcoef(X, y1)
```

```python
# delta系数取值数组
dl = [0, 0.5, 1, 1.5, 2, 5]

# 空列表容器
yl = []          # 不同delta下y的取值所组成的列表
cl = []          # 不同y下相关系数矩阵所组成的列表

# 计算不同delta下y和相关系数计算情况
for i in dl:
    yn = X + 1 + (ran * i)
    cl.append(np.corrcoef(X, yn))
    yl.append(yn)

# plt.plot(X, yl[0], 'o')
```

```python
for i in range(len(yl)):
    loc = 231 + i
    plt.subplot(loc)
    plt.plot(X, yl[i], 'o')
    plt.plot(X, y, 'r-')
```
